```python
import xmltodict
import pandas as pd
def xml_to_csv(xml, filename):
    obj = xml['annotation']['object']
    name = obj['name']
    xmin = obj['bndbox']['xmin']
    ymin = obj['bndbox']['ymin']
    xmax = obj['bndbox']['xmax']
    ymax = obj['bndbox']['ymax']
    size = xml['annotation']['size']
    width = size['width']
    height = size['height']
    csv = [name, filename, height, width, xmax, xmin, ymax, ymin]
    return csv

def parse(model, result):
    number = 1
    max_number = 102
    for i in range(number, max_number):
        filename = model + str(i) + '.xml'
        fin = open('Dataset/annotations/xmls/' + filename, 'r', encoding='utf8')
        xml = fin.read()
        fin.close()
        parsedxml = xmltodict.parse(xml)
        result.append(xml_to_csv(parsedxml, filename))
    return result[:-1]

model_1 = 'BB-8_'
model_2 = 'R2-D2_'
result = [['name','filename','height','width','xmax','xmin','ymax','ymin']]
parse(model_1, result)
parse(model_2, result)
```

```python
frame = pd.DataFrame(result) # собираем фрейм
frame.to_csv('all_droids.csv',index=False) #экспортируем в файл
```
